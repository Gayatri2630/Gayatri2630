```python
import pandas as pd
```

```python
df=pd.read_csv("/content/data_YesBank_StockPrices.csv")
```

```python
df
```

```python
# prompt: from the data in the given file give me the linear regression  between High and Low in python and also represent the title in that graph

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Extract 'High' and 'Low' columns
X = df['High'].values.reshape(-1, 1)
y = df['Low'].values

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the 'Low' values using the fitted model
y_pred = model.predict(X)

# Plot the data points and the regression line
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('High')
plt.ylabel('Low')
plt.title('Linear Regression: High vs Low')
plt.legend()
plt.show()

```
